Selection sort

```python
def selectionsort(arr):
    n=len(arr)
    for i in range(0,n-1):
        min_ind = i 
        for j in range(i+1,n):
            if(arr[j]<arr[min_ind]):
                min_ind = j
        arr[min_ind],arr[i] = arr[i] , arr[min_ind]


arr = [1,6,3,2,10,9]

selectionsort(arr)

print(arr)
```

Insertion Sort 

```python
def insertionsort(arr):
    n = len(arr)
    for i in range(1,n):
        key = arr[i]
        j=i-1
        while(j>=0 and key <arr[j]):
            arr[j+1] = arr[j]
            j = j-1
        arr[j+1] = key

    
    arr = [9,8,6,1,2,5]
    insertionsort(arr)
    print(arr)
```
